```python
import csv
import spacy


nlp = spacy.load("en_core_web_sm")


all_courses_dict = {
    'Programming Fundamentals': 'Computing – Core',
    'Object Oriented Programming': 'Computing – Core',
    'Data Structures & Algorithms': 'Computing – Core',
    'Discrete Structures': 'Computing – Core',
    'Operating Systems': 'Computing – Core',
    'Database Systems': 'Computing – Core',
    'Software Engineering': 'Computing – Core',
    'Computer Networks': 'Computing – Core',
    'Information Security': 'Computing – Core',
    'Final Year Project': 'Computing – Core',
    'English Composition & Comprehension': 'General Education',
    'Technical & Business Writing': 'General Education',
    'Communication & Presentation Skills': 'General Education',
    'Professional Practices': 'General Education',
    'Intro to Info. & Comm. Technologies': 'General Education',
    'Pakistan Studies': 'General Education',
    'Islamic Studies/Ethics': 'General Education',
    'Foreign Language': 'University Electives',
    'Social Service': 'University Electives',
    'Management Related': 'University Electives',
    'Social Science Related': 'University Electives',
    'Economy Related': 'University Electives',
    'Calculus & Analytical Geometry': 'Mathematics & Science Foundation',
    'Probability & Statistics': 'Mathematics & Science Foundation',
    'Linear Algebra': 'Mathematics & Science Foundation',
    'Applied Physics': 'Mathematics & Science Foundation',
    'Compiler Construction': 'Domain CS Core',
    'Comp. Organization & Assembly Language': 'Domain CS Core',
    'Digital Logic Design': 'Domain CS Core',
    'Design & Analysis of Algorithms': 'Domain CS Core',
    'Parallel & Distributed Computing': 'Domain CS Core',
    'Artificial Intelligence': 'Domain CS Core',
    'Theory of Automata': 'Domain CS Core',
    'Differential Equations': 'Domain CS Supporting',
    'Multi-variate Calculus': 'Domain CS Supporting',
    'Graph Theory': 'Domain CS Supporting',
    'Theory of Programming Languages': 'Domain CS Supporting',
    'Numerical Computing': 'Domain CS Supporting',
    'CS Elective – 1': 'Domain CS Electives ',
    'CS Elective – 2': 'Domain CS Electives ',
    'CS Elective – 3': 'Domain CS Electives ',
    'CS Elective – 4': 'Domain CS Electives ',
    'CS Elective – 5': 'Domain CS Electives '
}


def find_closest_match(course_name):
    max_similarity = 0
    best_match = None

    for key in all_courses_dict.keys():
        similarity = nlp(course_name).similarity(nlp(key))
        if similarity > max_similarity:
            max_similarity = similarity
            best_match = key

    return best_match


new_csv_filename = '/content/Std-002.csv'
new_courses = []

with open(new_csv_filename, 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        course_name = row['courseName']

        if course_name in all_courses_dict:
            category = all_courses_dict[course_name]
        else:

            matched_category = find_closest_match(course_name)
            category = all_courses_dict.get(matched_category, 'Unknown Category')


            all_courses_dict[course_name] = category
        new_courses.append((course_name, category))


print('Matched Courses and Categories:')
for course_name, category in new_courses:
    print(f'{course_name}: {category}')


total_courses = len(new_courses)
category_counts = {category: new_courses.count((course, category)) for course, category in new_courses}


print('Percentage for Each Category:')
for category, count in category_counts.items():
    percentage = count / total_courses * 100
    print(f'{category}: {percentage:.2f}%')


common_courses_percentage = sum(category_counts.get(category, 0) for category in ['General Education', 'University Electives', 'Mathematics & Science Foundation', 'Computing – Core']) / total_courses * 100

domain_courses_percentage = sum(category_counts.get(category, 0) for category in ['Domain CS Core', 'Domain CS Electives ', 'Domain CS Supporting']) / total_courses * 100


print(f'\nCommon Courses Percentage: {common_courses_percentage:.2f}%')
print(f'Domain Courses Percentage: {domain_courses_percentage:.2f}%')
if (common_courses_percentage >= 63) and (domain_courses_percentage >= 37):
      print('You passed!')
else:
      print('You did not pass.')
```
